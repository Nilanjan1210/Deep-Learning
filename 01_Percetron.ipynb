# Perceptron Model

The **Perceptron Model** is one of the earliest and simplest types of artificial neural networks, introduced by **Frank Rosenblatt in 1958**. It was designed to mimic the way biological neurons work.

---

## Structure of a Perceptron

A perceptron takes multiple inputs, applies **weights** to them, sums them up, and passes the result through an **activation function** to produce an output.

$
      y = f\left(\sum_{i=1}^{n} w_i x_i + b\right)
$

---

## How a Perceptron Works

1. **Input Processing**  
   - The perceptron receives multiple inputs, each multiplied by its corresponding weight.  

2. **Weighted Sum and Bias**  
   - These weighted inputs are summed together, and the bias term is added to this sum.  

3. **Activation Function**  
   - The resulting sum is passed through an activation function, which determines the final **binary output** based on a predefined threshold.  

4. **Learning (Training)**  
   - The perceptron learns by adjusting its weights and bias based on the error between its predicted output and the actual target output.  This process typically involves a **learning rule**, such as the **Perceptron Learning Rule**, where weights are updated to minimize errors.  

---

## ðŸ”¹ Limitations

- A single-layer perceptron can only solve **linearly separable problems** (like **AND, OR** logic gates).  
- It **cannot solve non-linear problems** (like **XOR**).  

---

## ðŸ”¹ Importance

- It is the **foundation of modern deep learning models**.  
- **Multilayer Perceptrons (MLPs)** and **Deep Neural Networks** evolved from this simple idea.  

---

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_classification
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

import warnings
```

```python
# Create Dataset
X, y = make_classification(
    n_samples=1000,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    random_state=21
)
```

```python
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)
plt.show()
```

```python
x_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)
```

```python
perceptron = Perceptron()
perceptron.fit(x_train, y_train)
```

```python
predict = perceptron.predict(X_test)

print("My accuracy score is",accuracy_score(y_test, predict))
```

```python
# we see also weights and bias
print("Weights are",perceptron.coef_)
print("Bias is",perceptron.intercept_)


print("Number of weight updates performed during training :", perceptron.t_)
print("The actual number of iterations to reach the stopping criterion :", perceptron.n_iter_)
print("Number of features seen during fit :", perceptron.n_features_in_)
print("The unique classes labels :", perceptron.classes_)
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(X=X, y=y, clf=perceptron)
plt.show()
```
